```python
#   (        21    ).
#      2       .
#    ( . .).
#       ().
#      (/).
#    (  /(  )^2).
# ,      (diabetes pedegree).
#  ().

from tensorflow.keras.models import Sequential # 
from tensorflow.keras.layers import Dense # 
import numpy

dataset = numpy.loadtxt("prima-indians-diabetes.csv",delimiter = ",")
X,Y = dataset[:,0:8],dataset[:,8]#   
# 
model = Sequential()
#  (5  )
model.add(Dense(12,input_dim=8, activation = 'relu'))#8 -      12 - -    
model.add(Dense(15,activation = 'relu'))#15    
model.add(Dense(8,activation = 'relu'))#8

model.add(Dense(10,activation = 'relu'))#
model.add(Dense(1,activation = 'sigmoid'))#     
                                    #       
model.compile(loss = "binary_crossentropy", optimizer = "adam", metrics = ["accuracy"])#  
                                                                                    # ,  
    
    
    
#
model.fit(X,Y,epochs = 1000, batch_size=10)# ,   ,    ,   (    10 )
#  
scores = model.evaluate(X,Y)# 2 :  -   ,  - 

#  evaluate()         :
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
print(scores[1])
```

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy

#    
numpy.random.seed(2)

#  ,        
dataset = numpy.loadtxt("prima-indians-diabetes.csv", delimiter=",")
#      (X)     (Y)
X, Y = dataset[:,0:8], dataset[:,8]

#  ,     
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu')) #     input_dim
model.add(Dense(15, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid')) #   relu   

#  ,    adam
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=['accuracy'])

#   
model.fit(X, Y, epochs = 1000, batch_size=10)

#  
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
